```python
import numpy as np 
import pandas as pd
import math
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

sns.set(rc={'figure.figsize':(12, 8)})
sns.set_style('ticks')
sns.set_palette("deep")
```

```python
df_oct = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2019-Oct_final.csv", index_col=0)

df_nov = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2019-Nov_final.csv", index_col=0)

df_dec = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2019-Dec_final.csv", index_col=0)

df_jan = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Jan_final.csv", index_col=0)

df_feb = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Feb_final.csv", index_col=0)

df_mar = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Mar_final.csv", index_col=0)

df_apr = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Apr_final.csv", index_col=0)
```

Połączenie zbiorów danych

```python
# połączenie dataframeów
frames = [df_oct, df_nov, df_dec, df_jan, df_feb, df_mar, df_apr]
df = pd.concat(frames).reset_index(drop=True)

df
```

```python
df.info(show_counts=True)
```

Sprawdzenie ciągłości dataframe'u

```python
sns.scatterplot(y="event_month", x="event_monthday", data=df)
plt.show()
```

Zapisanie dataframe'u
