# Przygotowanie zbiorów dla każego miesiąca zawierających tylko `event_type == purchase` do analizy RFM.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
from datetime import datetime
```

```python
cols = list(pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Apr_clean.csv", nrows =1))
cols.remove("category_id")
chunk_list = []
chunks = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/archive/2020-Apr_clean.csv", index_col=0, usecols=cols, chunksize=200000000)
for chunk in chunks:
    chunk = chunk.loc[chunk['event_type'] == 'purchase']
    chunk_list.append(chunk)

df = pd.concat(chunk_list)
```

```python
df.reset_index(inplace=True, drop=True)
df.info(show_counts=True)
```
